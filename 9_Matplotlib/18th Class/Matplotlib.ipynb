```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1,10)
y = np.tan(x)
```

```python
x
```

```python
y
```

## Line plot

```python
plt.figure(num = 1,figsize=(8,5))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("tan(y)")
plt.title("x vs tan(y)")
plt.show
```

```python
x = np.arange(1,10)
y = np.tan(x)
z = np.sin(y)
```

```python
x
```

```python
y
```

```python
z
```

```python
plt.figure(num=1,figsize=(8,5),dpi = 300)
plt.plot(x,y,label = "Tan Graph",color='red',linewidth=2, marker='o',linestyle='dashed',markersize = 15)
plt.plot(x,z,label = "Sin Graph",color = 'blue',linewidth=2, marker='^',markersize = 15)
plt.title("x vs wives")
plt.legend()
plt.grid(True)
# plt.xlim([2,8])
# plt.ylim([-2,0])
plt.show()
```

## Histogram

```python
data = y*z
plt.figure(num=1,figsize=(8,5))
plt.hist(data,bins=5,label='Tan Distance', orientation='vertical',color = 'red')
plt.title("Bar Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

## Bar Chart

```python
lang = ['Python','Jave','C++','R','Rubi','JavaScripts']
rating = [8.9,4.6,8.0,5.3,4.3,6.7]

plt.figure(num=1,figsize=(8,5))
plt.bar(lang,rating,color = 'blue',label = "Language Color")
plt.title("Language Rating")
plt.xlabel('Language')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.show()
```

```python
lang = ['Python','Jave','C++','R','Rubi','JavaScripts']
rating = [8.9,4.6,8.0,5.3,4.3,6.7]

plt.figure(num=1,figsize=(8,5))
plt.barh(lang,rating,color = 'blue',label = "Language Color")
plt.title("Language Rating")
plt.xlabel('Language')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.show()
```

```python
lang = ['Python','Jave','C++','Rubi','JavaScripts']
rating = [8.9,4.6,8.0,4.3,6.7]
color = ['lightgreen','gray','darkorange','lightblue','red',]

plt.figure(num=1,figsize=(6,8))
plt.pie(rating,labels=lang,colors = color, autopct="%.1f",explode=(0,0,0,0,0.1),shadow=True)
plt.title("Language Rating")
plt.xlabel('Language')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
plt.figure(num=1,figsize=(8,4))
plt.scatter(x,y,label = "Tan Graph",color='red',linewidth=2, marker='o')
plt.scatter(x,z,label = "Sin Graph",color = 'blue',linewidth=2, marker='^')
plt.title("x vs wives")
plt.legend()
plt.grid(True)
# plt.xlim([2,8])
# plt.ylim([-2,0])
plt.show()
```

# 3D Vizualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

## Line Plot 3D

```python
axes = plt.axes(projection = '3d')
axes.plot3D(x,y,z, label='XYZ')
axes.plot3D(x,y,z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

## Scatter Plot 3D

```python
axes = plt.axes(projection = '3d')
axes.scatter3D(x,y,z, label='XYZ')
axes.scatter3D(x,y,z**2, label='XYZ^2')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')
plt.legend()
plt.show()
```

## Multiplots

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y)
axes[0,0].set_title('Line plot')
axes[0,1].scatter(x,y)
axes[1,0].hist(data)
axes[1,1].bar(lang,rating)
plt.tight_layout()
```
