# colors transformation

```python
from skimage import color
```

```python
image = io.imread('bird.jpg', as_gray=True)
image
```

```python
io.imshow(image)
```

```python
image = io.imread('bird.jpg')
io.imshow(image)
```

```python
# color conversion
hsv_img = color.rgb2hsv(image)
io.imshow(hsv_img, cmap='hsv')
```

```python
rgb_img = color.hsv2rgb(hsv_img)
io.imshow(rgb_img)
```

```python
plt.imshow(hsv_img[:, :, 0], cmap='hsv')
```

```python
plt.imshow(hsv_img[:, :, 1], cmap='hsv')
```

```python
plt.imshow(hsv_img[:, :, 2], cmap='hsv')
```

# shape transformation

```python
from skimage import transform
```

```python
image.shape
```

```python
resized_img = transform.resize(image, output_shape=(64, 64))
resized_img.shape
```

```python
io.imshow(resized_img)
```

```python
rescaled_img = transform.rescale(image, 1/3, multichannel=True)
rescaled_img.shape
```

```python
io.imshow(rescaled_img)
```

```python
rotate_img = transform.rotate(image, 99, resize=True)
io.imshow(rotate_img)
```

# exposure transformation

```python
from skimage import exposure
```

```python
sig_img = exposure.adjust_sigmoid(image, 0.1)
io.imshow(sig_img)
```

```python
log_img = exposure.adjust_log(image, 0.9)
io.imshow(log_img)
```

```python
gamma_img = exposure.adjust_gamma(image, 1.5)
io.imshow(gamma_img)
```

# filters

```python
from skimage import filters
```

```python
f_img = filters.gaussian(image, 0.5, multichannel=True)
io.imshow(f_img)
```

# feature transformation

```python
from skimage import feature
```

```python
vector, img = feature.hog(image, channel_axis=-1, visualize=True)
vector.shape
```
